```python
#Python OOPs Questions
'''
1.What is Object-Oriented Programming (OOP)?
Ans. Object-Oriented Programming (OOP) is a programming paradigm that organizes
software design around objects rather than functions or logic. An object represents
a real-world entity and contains both data (attributes) and methods (functions) that operate on that data.
In OOP, a program is built by creating and interacting with objects, which makes
the code more modular, reusable, scalable, and easier to maintain.

Key Concepts of OOP
1. Class
A class is a blueprint or template for creating objects.
It defines the properties (variables) and behaviors (methods).
Example:
A Student class may have attributes like name, roll_no and methods like study().

2. Object
An object is an instance of a class.
It represents a real-world entity created using a class.
Example:
student1 is an object of the Student class.

3. Encapsulation
Encapsulation means wrapping data and methods together into a single unit (class) and restricting direct access to some data.
Benefit: Data security and controlled access.

4. Abstraction
Abstraction means hiding internal implementation details and showing only essential features to the user.
Example:
When you drive a car, you use the steering and pedals without knowing how the engine works.

5. Inheritance
Inheritance allows a class to acquire properties and methods of another class.
Benefit: Code reusability.
Example:
Teacher class can inherit from Person class.

6. Polymorphism
Polymorphism means one interface, multiple forms.
The same method name can perform different tasks depending on the object.
Example:
A method draw() behaves differently for Circle and Rectangle.

Advantages of OOP
Improves code reusability
Makes programs easier to understand and maintain
Enhances data security
Supports scalability for large projects

Languages That Support OOP
Python
Java
C++
C#
Ruby


2.What is a class in OOP?
Ans. A class is a blueprint or template used to create objects in Object-Oriented Programming.
It defines what an object will have (attributes/data) and what an object can do (methods/functions).
A class itself does not occupy memory until an object is created from it.
Example:
lass Student:
    def __init__(self, name, roll_no):
        self.name = name
        self.roll_no = roll_no

    def display(self):
        print("Name:", self.name)
        print("Roll No:", self.roll_no)

# Creating objects
s1 = Student("Amit", 101)
s2 = Student("Riya", 102)

s1.display()
s2.display()


3.What is an object in OOP?
Ans. An object is a real-world entity and an instance of a class.
It represents a specific implementation of a class and contains data (attributes) and behavior (methods).
Objects occupy memory, unlike classes.
Example:
s1 and s2 are objects of the Student class.


4.What is the difference between abstraction and encapsulation?
Ans. Abstraction:
Hides internal implementation details.
Shows only essential features to the user.
Focuses on what an object does.
Reduces complexity of the system.
Achieved using abstract classes and interfaces.

Encapsulation:
Wraps data and methods into a single unit (class).
Restricts direct access to data.
Focuses on how data is protected.
Improves data security and integrity.
Achieved using classes and access specifiers.


5.What are dunder methods in Python?
Ans. Dunder methods are special methods in Python whose names begin and end with double underscores (__).
The word dunder means double underscore. These methods allow developers to define the behavior of objects
when built-in functions or operators are used on them.

Key Points:
They are automatically invoked by Python.
They help customize object behavior.
They support operator overloading.
They make user-defined objects act like built-in data types.
Commonly used in object-oriented programming.

Examples:
__init__ : initializes an object
__str__ : returns a readable string representation
__len__ : returns the length of an object
__add__ : defines behavior of + operator


6.Explain the concept of inheritance in OOP.
Ans. Inheritance is a concept in OOP where one class (called the child or derived class) acquires
the properties and methods of another class (called the parent or base class). It helps in building new
classes using existing ones, reducing code duplication and improving reusability.

Key Points:
The child class can use and extend the features of the parent class.
It promotes code reusability.
It supports hierarchical relationships between classes.
Existing code does not need to be rewritten.
It improves maintainability of programs.

Example:
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):   # Dog inherits Animal
    def bark(self):
        print("Dog barks")

d = Dog()
d.speak()   # inherited method
d.bark()    # own method


7.What is polymorphism in OOP?
Ans. Polymorphism means many forms. In OOP, it refers to the ability of the same method or function name
to perform different tasks depending on the object that calls it. It allows flexibility and dynamic behavior in programs.

Key Points:
The same method name can have different implementations.
Behavior depends on the object type.
It improves code flexibility and scalability.
It supports method overriding and method overloading.
It is an important feature of OOP.

Example:
class Bird:
    def sound(self):
        print("Bird makes a sound")

class Sparrow(Bird):
    def sound(self):
        print("Sparrow chirps")

class Crow(Bird):
    def sound(self):
        print("Crow caws")

b1 = Sparrow()
b2 = Crow()

b1.sound()
b2.sound()


8.How is encapsulation achieved in Python?
Ans. Encapsulation is the concept of wrapping data and methods together and restricting direct access
to some of the object’s components to protect it from unauthorized changes.
In Python, it is mainly achieved using access specifiers.
Example:
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # private variable

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance

acc = BankAccount(1000)
acc.deposit(500)
print(acc.get_balance())  # 1500
# print(acc.__balance)    # Error: Cannot access private variable


9.What is a constructor in Python?
Ans. A constructor is a special method in Python used to initialize objects of a class. It is called
automatically when an object is created. In Python, the constructor is always named __init__().

Key Points:
Initializes object attributes.
Automatically called when an object is created.
Can accept parameters to set initial values.
Helps in setting up the object for use.

Example:
class Student:
    def __init__(self, name, roll_no):
        self.name = name
        self.roll_no = roll_no

s1 = Student("Amit", 101)
print(s1.name)      # Amit
print(s1.roll_no)   # 101


10.What are class and static methods in Python?
Ans. In Python, class methods and static methods are special types of methods
that are not tied to individual objects like regular instance methods. They are used for different purposes.

Class Method:
Defined using the @classmethod decorator.
Takes cls as the first parameter (refers to the class).
Can access or modify class variables but not instance variables.

Example:
class Student:
    school = "ABC School"
    @classmethod
    def get_school(cls):
        print("School:", cls.school)
Student.get_school()  # School: ABC School

Static Method:
Defined using the @staticmethod decorator.
Does not take self or cls as a parameter.
Cannot access instance or class variables directly.
Used for utility functions related to the class.

Example:
class Math:
    @staticmethod
    def add(a, b):
        return a + b
print(Math.add(5, 3))  # 8


11.What is method overloading in Python?
Ans. Method overloading is the ability to define multiple methods with the same name but different parameters in a class.
It allows a method to behave differently based on the number or type of arguments.
Key Points:
Python does not support traditional method overloading like Java or C++.
Can be simulated using default arguments or *args.
Helps make methods flexible for different inputs.

Example:
class Math:
    def add(self, a, b=0):
        return a + b
m = Math()
print(m.add(5, 3))  # 8
print(m.add(5))     # 5

In Python, method overloading is achieved using default parameters or variable-length arguments.


12. What is method overriding in OOP?
Ans. Method overriding occurs when a child class provides its own implementation of a method that is already defined in the
parent class. It allows a child class to change or extend the behavior of inherited methods.

Key Points:
The method in the child class has the same name, parameters, and signature as in the parent class.
Enables runtime polymorphism.
Used to customize or extend the behavior of inherited methods.
The parent class method can still be called using super().

Example:
class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def sound(self):       # overriding method
        print("Dog barks")

d = Dog()
d.sound()   # Dog barks


13. What is a property decorator in Python?
Ans. property decorator (@property) is used to access a method like an attribute. It allows controlled access to
instance variables without directly accessing them, combining encapsulation and readability.

Key Points:
Used to get the value of a private attribute.
Can be combined with setter (@attribute_name.setter) to update the value.
Helps in data validation while accessing or modifying attributes.

Example (Python):
class Student:
    def __init__(self, name):
        self.__name = name  # private attribute

    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, value):
        if value != "":
            self.__name = value
        else:
            print("Name cannot be empty")

s = Student("Amit")
print(s.name)   # Amit
s.name = "Riya"
print(s.name)   # Riya


14.Why is polymorphism important in OOP?
Ans. Polymorphism is important because it allows objects of different classes to be treated in a uniform way,
making programs more flexible and scalable. It helps in reducing code complexity and supports dynamic behavior.
Enables one interface, multiple implementations.
Promotes code reusability and maintainability.
Supports runtime flexibility with method overriding.
Makes programs easier to extend without changing existing code.
Helps in designing generic and modular code.


15.What is an abstract class in Python?
Ans. An abstract class is a class that cannot be instantiated and is meant to be inherited by other classes.
It is used to define common methods that must be implemented by its child classes.

Key Points:
Defined using the abc module and @abstractmethod decorator.
Can have abstract methods (without implementation) and regular methods (with implementation).
Ensures that all child classes implement the abstract methods.
Helps in achieving abstraction and designing a template for other classes.

Example (Python):
from abc import ABC, abstractmethod
class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass  # abstract method

class Dog(Animal):
    def sound(self):
        print("Dog barks")

d = Dog()
d.sound()  # Dog barks


16.What are the advantages of OOP?
Ans. Code Reusability: Classes and objects can be reused across programs.
Modularity: Programs are divided into independent classes, making them easier to manage.
Data Security: Encapsulation protects data from unauthorized access.
Flexibility: Polymorphism and inheritance allow dynamic behavior and code extension.
Ease of Maintenance: Changes in one class don’t affect other parts of the program.
Real-world Modeling: Objects represent real-world entities, making programs intuitive.


17.What is the difference between a class variable and an instance variable?
Ans. In Python, variables in a class can be either class variables or instance variables, and they behave differently.

Class Variable:
Shared by all objects of the class.
Defined inside the class but outside any method.
Changes affect all instances of the class.

Instance Variable:
Unique to each object.
Defined inside the constructor (__init__) using self).
Changes affect only that specific object.

Example:
class Student:
    school = "ABC School"  # class variable

    def __init__(self, name):
        self.name = name   # instance variable

s1 = Student("Amit")
s2 = Student("Riya")

print(s1.school, s1.name)  # ABC School Amit
print(s2.school, s2.name)  # ABC School Riya

Student.school = "XYZ School"
print(s1.school, s2.school)  # XYZ School XYZ School


18.What is multiple inheritance in Python?
Ans. Multiple inheritance is a feature in Python where a child class can inherit from more than one parent class.
This allows the child class to access attributes and methods from multiple classes.

Key Points:
Enables reusing code from multiple classes.
Can create complex relationships between classes.
Python uses Method Resolution Order (MRO) to determine which parent’s method is called if there’s a conflict.
Helps in building flexible and modular programs.

Example:
class Father:
    def skills(self):
        print("Gardening")

class Mother:
    def skills(self):
        print("Cooking")

class Child(Father, Mother):
    pass

c = Child()
c.skills()  # Gardening (Father's method is called first due to MRO)


19. Explain the purpose of ‘’__str__’ and ‘__repr__’ ‘ methods in Python.
Ans. These are dunder methods used to provide string representations of objects, but they serve slightly different purposes.

__str__
Returns a readable, user-friendly string of the object.
Used by print() and str() functions.
Meant for end-users.

__repr__
Returns an unambiguous string that can be used to recreate the object.
Used by repr() function and in the interpreter.
Meant for developers/debugging.

Example:
class Student:
    def __init__(self, name, roll):
        self.name = name
        self.roll = roll

    def __str__(self):
        return f"Student Name: {self.name}"

    def __repr__(self):
        return f"Student('{self.name}', {self.roll})"

s = Student("Amit", 101)
print(s)      # uses __str__: Student Name: Amit
print(repr(s)) # uses __repr__: Student('Amit', 101)


20.What is the significance of the ‘super()’ function in Python?
Ans. The super() function is used to call a method from a parent (base) class inside a child (derived) class.
It helps in reusing and extending functionality without rewriting code.

Key Points:
Accesses parent class methods and constructors.
Commonly used in method overriding to extend parent behavior.
Helps avoid explicitly naming the parent class, making code maintainable.
Supports multiple inheritance by following Python’s Method Resolution Order (MRO).

Example:
class Parent:
    def greet(self):
        print("Hello from Parent")

class Child(Parent):
    def greet(self):
        super().greet()  # calls Parent's greet
        print("Hello from Child")

c = Child()
c.greet()

Output:
Hello from Parent
Hello from Child


21.What is the significance of the __del__ method in Python?
Ans. The __del__ method is a destructor in Python. It is automatically called when an object is about to be destroyed
or garbage collected. Its main purpose is to clean up resources before the object is removed from memory.

Key Points:
Used to release resources like files, network connections, or database connections.
Called automatically when reference count of an object becomes zero.
Helps in preventing memory leaks.
Rarely needed in Python due to automatic garbage collection.

Example:
class FileHandler:
    def __init__(self, filename):
        self.file = open(filename, "w")
        print("File opened")

    def __del__(self):
        self.file.close()
        print("File closed")

f = FileHandler("test.txt")
del f  # __del__ is called automatically


22. What is the difference between @staticmethod and @classmethod in Python?
Ans. In Python, both are decorators used to define methods that are not regular instance methods, but they behave differently.

@staticmethod
Does not take self or cls as the first parameter.
Cannot access instance variables or class variables directly.
Used for utility functions related to the class.

@classmethod
Takes cls as the first parameter (refers to the class).
Can access class variables and other class methods.
Used to modify class state or create alternative constructors.

Example:
class Student:
    school = "ABC School"

    @classmethod
    def get_school(cls):
        print("School:", cls.school)

    @staticmethod
    def greet():
        print("Welcome to the school!")

Student.get_school()  # ABC School
Student.greet()       # Welcome to the school!


23.How does polymorphism work in Python with inheritance?
Ans. Polymorphism allows objects of different classes to be treated uniformly through a common interface, especially
when classes share a parent-child relationship. When a child class overrides a parent’s method, Python decides at
runtime which method to call based on the object. This is called runtime polymorphism or method overriding.

Key Points:
Enables one method name to behave differently for different objects.
Works naturally with inheritance.
Promotes code flexibility and reusability.
Supports dynamic method dispatch in Python.

Example:

class Animal:
    def sound(self):
        print("Some sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

# Using polymorphism
def make_sound(animal):
    animal.sound()

d = Dog()
c = Cat()
make_sound(d)  # Dog barks
make_sound(c)  # Cat meows


24.What is method chaining in Python OOP?
Ans. Method chaining is a technique in Python where multiple methods are called on the same object in a single line.
Each method returns the object itself (self), allowing a chain of method calls. This makes code concise and readable.

Key Points:
Methods must return self to continue the chain.
Improves code readability and flow.
Commonly used in fluent interfaces or libraries like Pandas.

Example:
class Student:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print(f"Hello {self.name}")
        return self

    def study(self):
        print(f"{self.name} is studying")
        return self

s = Student("Amit")
s.greet().study()

Output:
Hello Amit
Amit is studying


25.What is the purpose of the __call__ method in Python?
Ans. The __call__ method allows an object of a class to be called like a function. By defining __call__ in a class,
you can make instances behave like callable functions, which is useful for flexible and intuitive object behavior.

Key Points:
Makes an object callable using parentheses ().
Can accept arguments just like a normal function.
Useful for functional programming or creating function-like objects.
Helps encapsulate behavior inside an object while keeping function syntax.

Example (Python):
class Adder:
    def __init__(self, n):
        self.n = n

    def __call__(self, x):
        return self.n + x

add_five = Adder(5)
print(add_five(10))  # 15
'''
```

```python
#2.Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle
#from it and implement the area() method in both.
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

c = Circle(5)
r = Rectangle(4, 6)

print("Circle area:", c.area())
print("Rectangle area:", r.area())
```

```python
#5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
#balance and methods to deposit, withdraw, and check balance.
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: {amount}")
        else:
            print("Insufficient balance")

    def get_balance(self):
        return self.__balance

account = BankAccount(1000)
account.deposit(500)
account.withdraw(300)
print("Balance:", account.get_balance())
```

```python
#14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked
#and hourly rate. Create a derived class Manager that adds a bonus to the salary.
class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        return super().calculate_salary() + self.bonus

e = Employee("Amit", 40, 200)
m = Manager("Riya", 40, 200, 5000)

print("Employee Salary:", e.calculate_salary())
print("Manager Salary:", m.calculate_salary())
```
